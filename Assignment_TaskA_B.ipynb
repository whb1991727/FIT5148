```python
def parallel_outjoin(File1, File2, range_indices, key1, key2, processor):
    #start = time.time()
    datasetsL = range_partition(File1,range_indices,key1)
    datasetsR = range_partition(File2,range_indices,key2)
    #end = time.time()
    results = []
    
    LL = len(datasetsL)
    number = 0
    pool = mp.Pool(processes=processor)
   
    while number < LL:
        pool.apply_async(nl_leftjoin,(datasetsL[number],datasetsR[number],key1,key2),callback=results.extend) # Allocation happened
        number = number + 1
    # Searching will be triggered when pool.join or result.get is called
    pool.close()                                
    pool.join()
    
    #print(end-start)
    return results 
    
```
